## Quality
- year has many non-year values
- year object to int
- price has Ask for price
- Price object to int
- kms driven has kms with integers
- kms driven obbject to int
- kms driven has null values
- fuel_type has null values
- keep first 3 words of name

## Cleaning

```python
backup = df.copy()
```

```python
df = df[df["year"].str.isnumeric()]
```

```python
df["year"] = df["year"].astype(int)
```

```python
df.info()
```

```python
df = df[df["Price"] != "Ask For Price"]
```

```python
df["Price"] = df["Price"].str.replace(",", "").astype(int)
```

```python
df.info()
```

```python
df["kms_driven"] = df["kms_driven"].str.split(" ").str.get(0).str.replace(",", "")
```

```python
df = df[df["kms_driven"] != "Petrol"]
```

```python
df["kms_driven"] = df["kms_driven"].astype(int)
```

```python
df.info()
```

```python
df = df[~df["fuel_type"].isna()]
```

```python
df.info()
```

```python
df["name"] = df["name"].str.split(" ").str.slice(0,3).str.join(" ")
```

```python
df = df.reset_index(drop = True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df[df["Price"]< 6e6]
```

```python
df.reset_index(drop =True)
```

```python
final_df = df.copy()
```

```python
df.to_csv("Cleaned_car.csv")
```

## Model

```python
X = df.drop(columns ="Price")
y = df["Price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder(handle_unknown="ignore")
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(handle_unknown="ignore",categories= ohe.categories_),['name','company','fuel_type']),
                                      remainder="passthrough")
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred =pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X_train,y_train,cv=5,scoring ="r2")
```

```python
scores = cross_val_score(pipe, X_train,y_train,cv=5,scoring ="r2")
```

```python
print(scores.mean())
```

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)
```

```python
pipe_gb = make_pipeline(column_trans, model)
```

```python
pipe_gb.fit(X_train, y_train)
```

```python
y_pred_gb = pipe_gb.predict(X_test)
r2_score(y_test,y_pred_gb)
```

```python
scores = cross_val_score(pipe_gb, X_train,y_train, cv=5, scoring="r2")
```

```python
scores
```

```python
print("GB",scores.mean() )
```

```python
# Since our dataset is small, So we need to check for  many train test split in order to get max accuracy
```

```python
scores =[]
for i in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred = pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
max(scores)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```
